```python
import pandas as pd

# Load the dataset (assuming it's a CSV file)
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

fixed_acidity = df['fixed acidity']
volatile_acidity = df['volatile acidity']

plt.figure(figsize=(10, 8))  # Adjust the figure size as needed

plt.scatter(fixed_acidity, volatile_acidity, marker='o', c='blue', label='Scatter Plot')

plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.title('Scatter Plot of Fixed Acidity vs. Volatile Acidity')

plt.legend()

plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the data into X (independent variable) and y (dependent variable)
X = df[['pH']]
y = df['quality']

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Calculate the mean squared error
mse = mean_squared_error(y, y_pred)

# Round the predicted values to integers
y_pred_rounded = [round(val) for val in y_pred]

print("Original Quality values:", y.tolist())
print("Predicted Quality values:", y_pred_rounded)
print("Mean Squared Error:", mse)
```
